```python
import cv2
import numpy as np

def to_gray(frame):
    """Renkli görüntüyü gri seviyeye çevirme."""
    return cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

# Uzamsal filtreler (gürültü azaltma)
def mean_filter(gray, ksize=5):
    return cv2.blur(gray, (ksize, ksize))

def gaussian_filter(gray, ksize=5, sigma=0):
    return cv2.GaussianBlur(gray, (ksize, ksize), sigmaX=sigma)

def median_filter(gray, ksize=5):
    return cv2.medianBlur(gray, ksize)

def bilateral_filter(gray, d=9, sigma_color=75, sigma_space=75):
    return cv2.bilateralFilter(gray, d, sigma_color, sigma_space)

#  Kontrast işlemleri
def contrast_stretch(gray):
    g_min = np.min(gray)
    g_max = np.max(gray)
    if g_max - g_min == 0:
        return gray.copy()
    stretched = (gray - g_min) * (255.0 / (g_max - g_min))
    return stretched.astype(np.uint8)

def gamma_correction(gray, gamma=1.0):
    normalized = gray / 255.0
    corrected = np.power(normalized, gamma)
    corrected = np.clip(corrected * 255.0, 0, 255)
    return corrected.astype(np.uint8)

#  Histogram işlemleri
def hist_equalization(gray):
    return cv2.equalizeHist(gray)

def clahe_contrast(gray, clip_limit=2.0, tile_grid_size=(8, 8)):
    clahe = cv2.createCLAHE(clipLimit=clip_limit, tileGridSize=tile_grid_size)
    return clahe.apply(gray)

#  Kenar bulma
def sobel_edges(gray):
    sobel_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
    sobel_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)
    mag = np.sqrt(sobel_x**2 + sobel_y**2)
    mag = (mag / mag.max()) * 255
    return mag.astype(np.uint8)

def canny_edges(gray, t1=50, t2=150):
    return cv2.Canny(gray, t1, t2)

#  Morfolojik işlemler
def morph_ops(edge_img, op="dilate", ksize=3, iterations=1):
    kernel = np.ones((ksize, ksize), np.uint8)
    if op == "dilate":
        return cv2.dilate(edge_img, kernel, iterations=iterations)
    elif op == "erode":
        return cv2.erode(edge_img, kernel, iterations=iterations)
    elif op == "open":
        return cv2.morphologyEx(edge_img, cv2.MORPH_OPEN, kernel, iterations=iterations)
    elif op == "close":
        return cv2.morphologyEx(edge_img, cv2.MORPH_CLOSE, kernel, iterations=iterations)
    else:
        return edge_img
```

```python
video_path = "ornek.mp4"
cap = cv2.VideoCapture(video_path)
ret, frame = cap.read()
cap.release()

if not ret:
    raise RuntimeError("Video karesi okunamadı")

gray = to_gray(frame)

mean_img      = mean_filter(gray, 5)
gauss_img     = gaussian_filter(gray, 5)
median_img    = median_filter(gray, 5)
bilateral_img = bilateral_filter(gray)

cv2.imshow("Orijinal Gri", gray)
cv2.imshow("Mean", mean_img)
cv2.imshow("Gaussian", gauss_img)
cv2.imshow("Median", median_img)
cv2.imshow("Bilateral", bilateral_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
stretched    = contrast_stretch(gray)
gamma_light  = gamma_correction(gray, gamma=0.7)
gamma_dark   = gamma_correction(gray, gamma=1.5)
hist_eq      = hist_equalization(gray)
clahe_img    = clahe_contrast(gray, clip_limit=2.0, tile_grid_size=(8, 8))

cv2.imshow("Orijinal Gri", gray)
cv2.imshow("Kontrast Germe", stretched)
cv2.imshow("Gamma 0.7", gamma_light)
cv2.imshow("Gamma 1.5", gamma_dark)
cv2.imshow("Hist Esl.", hist_eq)
cv2.imshow("CLAHE", clahe_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
denoised  = median_filter(gray, 5)
enhanced  = clahe_contrast(denoised, clip_limit=2.0, tile_grid_size=(8, 8))

sobel_img      = sobel_edges(enhanced)
canny_normal   = canny_edges(enhanced, 50, 150)
canny_sensitif = canny_edges(enhanced, 30, 100)

cv2.imshow("Enhanced (Median + CLAHE)", enhanced)
cv2.imshow("Sobel Magnitude", sobel_img)
cv2.imshow("Canny 50-150", canny_normal)
cv2.imshow("Canny 30-100", canny_sensitif)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def preprocess_for_zebra(frame):
    gray = to_gray(frame)

    denoised = median_filter(gray, 5)
    enhanced = clahe_contrast(denoised, clip_limit=2.0, tile_grid_size=(8, 8))

    edges = canny_edges(enhanced, 50, 150)

    edges = morph_ops(edges, op="dilate", ksize=3, iterations=1)
    edges = morph_ops(edges, op="close", ksize=5, iterations=1)

    return edges
```

```python
def detect_lines(edge_img):
    lines = cv2.HoughLinesP(
        edge_img,
        rho=1,
        theta=np.pi/180,
        threshold=80,
        minLineLength=40,
        maxLineGap=20
    )
    return lines

def is_zebra(lines):
    if lines is None or len(lines) < 3:
        return False

    angles = []
    for x1, y1, x2, y2 in lines[:, 0]:
        angle = np.degrees(np.arctan2(y2 - y1, x2 - x1))
        angles.append(abs(angle))

    mean_angle = np.mean(angles)
    std_angle  = np.std(angles)

    if std_angle < 10:
        return True
    return False
```

```python
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Video açılamadı.")
else:
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        edges = preprocess_for_zebra(frame)
        lines = detect_lines(edges)

        line_count = 0
        zebra_found = False

        if lines is not None:
            line_count = len(lines)
            for x1, y1, x2, y2 in lines[:, 0]:
                cv2.line(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)

            zebra_found = is_zebra(lines)

        cv2.putText(frame, f"Cizgi: {line_count}",
                    (20, 40), cv2.FONT_HERSHEY_SIMPLEX,
                    1, (0, 255, 0), 2)

        if zebra_found:
            cv2.putText(frame, "YAYA GECIDI TESPIT EDILDI-",
                        (20, 80), cv2.FONT_HERSHEY_SIMPLEX,
                        1, (0, 0, 255), 3)

        cv2.imshow("Video + Zebra", frame)
        cv2.imshow("Edges (Gelismis)", edges)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```
